```python
def clean_lodes(df):
    df['RETAIL'] = df['CNS07']
    df['EDHEALTH'] = df['CNS15'] + df['CNS16']
    df['LEISURE'] = df['CNS17'] + df['CNS18']
    df['OTHER'] = df['C000'] - df['RETAIL'] - df['EDHEALTH'] - df['LEISURE']
```
