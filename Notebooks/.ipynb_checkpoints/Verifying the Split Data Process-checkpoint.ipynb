```python
import pandas as pd
pd.set_option('max_rows',500)
```

```python
tenth = pd.read_csv('E:/Transit-Casa-Share/Data/2010_Split_Buffers_Tenth.csv')
```

```python
quarter = pd.read_csv('E:/Transit-Casa-Share/Data/2010_Split_Buffers_Quarter.csv')
```

```python
third = pd.read_csv('E:/Transit-Casa-Share/Data/2010_Split_Buffers_Third.csv')
```

```python
# for a tenth mile radius the buffers should be 20.1 acres and there are 3,563 buffers
# meaning the total area should be 71,602 acres
# percent diff is 0.48%

area10 =  tenth[['SPLIT_AREA','STOP_ID']].groupby(by = 'STOP_ID').sum()
print(area10.head())
area10['SPLIT_AREA'] = area10.SPLIT_AREA.astype(int)
print('Buffer size = ' + str(area10.SPLIT_AREA.unique()))
sum(area10.SPLIT_AREA)
```

```python
# for a quarter mile radius the buffers should be 125.6 acres and there are 3,563 buffers
# meaning the total area should be 447,512 acres 
# percent diff is 0.4%
    
area4 =  quarter[['SPLIT_AREA','STOP_ID']].groupby(by = 'STOP_ID').sum()
print(area4.head())
area4['SPLIT_AREA'] = area4.SPLIT_AREA.astype(int)
print('Buffer size = ' + str(area4.SPLIT_AREA.unique()))
sum(area4.SPLIT_AREA)
```

```python
# for a quarter mile radius the buffers should be 223.3 acres and there are 3,563 buffers
# meaning the total area should be 795,577 acres 
# percent diff is 0.11%
    
area3 =  third[['SPLIT_AREA','STOP_ID']].groupby(by = 'STOP_ID').sum()
print(area3.head())
area4['SPLIT_AREA'] = area4.SPLIT_AREA.astype(int)
print('Buffer size = ' + str(area3.SPLIT_AREA.unique()))
sum(area3.SPLIT_AREA)
```
