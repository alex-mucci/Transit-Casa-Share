```python
import pandas as pd
import geopandas as gp
%matplotlib inline
pd.set_option('max_columns',500)
```

```python
gp.read_file?
```

```python
pop = gp.read_file('E:/Transit-Casa-Alex/Input/2010 Population (Blocks)/Population_Houses_Blocks.shp')
```

```python
emp = pd.read_csv('E:/Transit-Casa-Alex/Input/LEHD Census Blocks/2009/Combined Employment/Combined Employment.csv')
```

```python
pop = pop.to_crs({'init' : 'epsg:2227'})
```

```python
pop['ALAND10'] = pop['geometry'].area
```

```python
pop.BLOCKID10 = pop.BLOCKID10.astype(float)
```
