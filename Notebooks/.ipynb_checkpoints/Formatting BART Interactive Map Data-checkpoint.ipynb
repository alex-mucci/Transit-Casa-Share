```python
import pandas as pd
import geopandas as gp
```

```python
bart16 = pd.read_csv('E:/Transit-Casa-Alex/Output/BART/2016/BART_Intra_SF_Ridership.csv',thousands = ',')
bart16 = bart16.drop('Unnamed: 0',axis = 1)
```

```python
bart09 = pd.read_csv('E:/Transit-Casa-Alex/Output/BART/2009/BART_Intra_SF_Ridership.csv',thousands = ',')
bart09 = bart09.drop('Unnamed: 0',axis = 1)
```

```python
bart09.head()
```

```python
bart16.head()
```

```python
bart = pd.merge(bart09,bart16,on = ['Station','STATION','CO'], suffixes = ('_09','_16'))
```

```python
bart.head()
```

```python
bart['FROM_DIFF'] = bart.Froms_16.astype(int) - bart.Froms_09.astype(int)
bart['TO_DIFF'] = bart.Tos_16.astype(int) - bart.Tos_09.astype(int)
bart['FROM_P_DIFF'] = (bart['FROM_DIFF']/bart['Froms_09'])*100
bart['FROM_P_DIFF'] = bart.FROM_P_DIFF.astype(int)
bart['FROM_P_DIFF'] = bart['FROM_P_DIFF'].apply(lambda value: str(value) + '%')

bart['TO_P_DIFF'] = (bart['TO_DIFF']/bart['Tos_09'])*100
bart['TO_P_DIFF'] = bart['TO_P_DIFF'].astype(int)
bart['TO_P_DIFF'] = bart['TO_P_DIFF'].apply(lambda value: str(value) + '%')
```

```python
bart.head()
```

```python
stations = gp.read_file('E:/Transit-Casa-Alex/Input/BART/Stations/BART_Sta_13.shp')
```

```python
stations = stations[stations['CO'] == 'SF']
stations
```

```python
map_data = stations[['STATION','geometry']].merge(bart,on = 'STATION')
```

```python
map_data['LAT'] = [-122.396033220569,-122.4015398362278,-122.4068235887763,-122.41418453044345,-122.4194566417886,-122.418304517505,-122.4339599538094,-122.4471285869132]
```

```python
map_data['LON'] = [37.79322220278397,37.78888642284969,37.78476966169613,37.77896496103885,37.76468721127412,37.75278416743183,37.73313053773821,37.72160655217209]
```
