```python
import pandas as pd
import geopandas as gp
%matplotlib inline
pd.set_option('display.max_columns',1000)
```

```python
#read in data 
data = gp.read_file('E:/Transit-Casa-Alex/Output/Buffers/Tenth/Census Data Buffers/Census_Buffers.shp',crs = 'crs')
```

```python
drop = ['AWATER10', 'BLOCKCE10', 'BLOCKID10', 
       'COUNTYFP10', 'FID_1',
       'FUNCSTAT10', 'GEOID10', 'INTPTLAT10', 'INTPTLON10',
       'Join_Cou_1', 'Join_Count', 'MTFCC10',
       'NAME10', 'ORIG_FID', 'PARTFLG', 
       'STATEFP10', 'STOP_LAT',
       'STOP_LON', 'TARGET_FID', 'TARGET_F_1', 
       'TRACTCE10', 'UACE10', 'UATYP10', 'UR10', 'createda_1', 'createdate',
       'h_geocode', 'w_geocode']
```

```python
rename = ['ALAND10', 'BUFF_DIST',
       'COMP_STOPS', 'EDHEALTH_RAC', 'EDHEALTH_WAC', 
       'HOUSING10', 
       'LEISURE_RAC', 'LEISURE_WAC', 
       'OTHER_RAC', 'OTHER_WAC', 'POP10',
       'RETAIL_RAC', 'RETAIL_WAC', 'STOP_ID', 
       'TOTAL_EMP_RAC', 'TOTAL_EMP_WAC',
       'geometry']
```

```python
for column in drop:
    data = data.drop(column,axis = 1)
```

```python
data.columns = rename
```

```python
#Aaland is area in meters of the census blocks that were intersected with the buffer 
#normalize the data with Aaland and then scale it to per acre for convienance (its intuitive)
items = ['_RAC','_WAC']
for item in items:
    data['EMPLOY_DEN'+ item] = (data['TOTAL_EMP'+item] / data['ALAND10'])/0.000247105
    data['RETAIL_DEN'+ item] = (data['RETAIL'+item] / data['ALAND10'])/0.000247105
    data['LEISURE_DEN'+item] = (data['LEISURE'+item] / data['ALAND10'])/0.000247105
    data['OTHER_DEN'+item] = (data['OTHER'+item] / data['ALAND10'])/0.000247105
    data['EDHEALTH_DEN'+item] = (data['EDHEALTH'+item] / data['ALAND10'])/0.000247105
    data['HOUSING10_DEN'] = (data['HOUSING10'] / data['ALAND10'])/0.000247105
    data['POP10_DEN'] = (data['POP10'] / data['ALAND10'])/0.000247105
```

```python
drop = ['EDHEALTH_RAC', 'EDHEALTH_WAC', 'HOUSING10',
'LEISURE_RAC', 'LEISURE_WAC', 'OTHER_RAC', 'OTHER_WAC',
'POP10', 'RETAIL_RAC', 'RETAIL_WAC', 'TOTAL_EMP_RAC',
'TOTAL_EMP_WAC']
```

```python
for column in drop:
    data = data.drop(column,axis =1 )
```
