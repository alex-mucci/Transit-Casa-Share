```python
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as smf

pd.set_option('display.max_columns', 500)


%matplotlib inline
```

```python
estimate = pd.read_csv('estimate2.csv')
```

```python
estimate['FREQ_S'] = 1 / (estimate['HEADWAY_S'] / 60.0)
```

```python
estimate.TOTSPEED = estimate.TOTSPEED.clip(lower = 0, upper = 40)
```
