```python
import csv as csv
import pandas as pd
import geopandas as gp
%matplotlib inline
pd.set_option('display.height', 100)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 100)
```

```python
store = pd.HDFStore("E:/Transit_Casa/Output/sfmuni_monthly_ts.h5")

bus = store.get('stop_day')
bus = bus[bus["MONTH"].isin(pd.date_range('2009-01-01', '2009-12-01'))]
```

```python
data = gp.read_file('E:/Transit-Casa-Alex/Output/Modeling/Estimate Shapefile/Estimation_File.shp')
```

```python
estimation_file = pd.merge(bus,data, on = 'STOP_ID', how = 'left')
```

```python
estimation_file.head()
```

```python
estimation_file.columns
```

```python
drop = ['STOP_LAT_y', 'STOP_LON_y', 'TARGET_FID', 'TARGET_F_1', 'AREA_FT', 'BLOCKID10','Join_Cou_1']
```

```python
for column in drop:
    estimation_file.drop(column, axis = 1)
```

```python
estimate = gp.GeoDataFrame(estimation_file)
```

```python
estimate['MONTH'] = estimate.MONTH.astype(str)
```

```python
estimate.to_file('Buffer10_Data.shp',driver = 'ESRI Shapefile')
```

```python
estimation_file.to_csv('E:\Transit-Casa-Alex\Output\Modeling/Buffer10_Data.csv')
```
