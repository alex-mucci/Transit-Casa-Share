```python
import pandas as pd
import geopandas as gp
```

```python
qcew_sf_16 = pd.read_csv('E:/Transit-Casa-Alex/Input/QCEW/2016/2016.q1-q4 06075 San Francisco County, California.csv')
qcew_sf_14 = pd.read_csv('E:/Transit-Casa-Alex/Input/QCEW/2014/2014.q1-q4 06075 San Francisco County, California.csv')
```

```python
codes = ['10','44-45','61','62','71','72']
```

```python
qcew_sf2_16 = qcew_sf_16[qcew_sf_16['industry_code'].isin(codes)]
qcew_sf2_14 = qcew_sf_14[qcew_sf_14['industry_code'].isin(codes)]
```

```python
qcew_sf3_16 = qcew_sf2_16[qcew_sf2_16['qtr'] == 4]
qcew_sf3_14 = qcew_sf2_14[qcew_sf2_14['qtr'] == 4]
```

```python
#drop out the first row because it was a total and not needed
qcew_sf3_16 = qcew_sf3_16.ix[4:]
qcew_sf3_14 = qcew_sf3_14.ix[4:]
```

```python
grouped_16 = qcew_sf3_16[['industry_code','year','month1_emplvl','month2_emplvl','month3_emplvl']].groupby(by = 'industry_code',as_index = False).agg({'year':'first','month1_emplvl':'sum','month2_emplvl':'sum','month3_emplvl':'sum'})
grouped_14 = qcew_sf3_14[['industry_code','year','month1_emplvl','month2_emplvl','month3_emplvl']].groupby(by = 'industry_code',as_index = False).agg({'year':'first','month1_emplvl':'sum','month2_emplvl':'sum','month3_emplvl':'sum'})
```

```python
grouped_16['Q4_AVG'] = (grouped_16['month1_emplvl'] + grouped_16['month2_emplvl'] + grouped_16['month3_emplvl'])/3
grouped_14['Q4_AVG'] = (grouped_14['month1_emplvl'] + grouped_14['month2_emplvl'] + grouped_14['month3_emplvl'])/3
```

```python
grouped_16.to_csv('E:\Transit-Casa-Alex\Output\QCEW/SF_Quarter4_Annual_2016.csv')
grouped_14.to_csv('E:\Transit-Casa-Alex\Output\QCEW/SF_Quarter4_Annual_2014.csv')
```
