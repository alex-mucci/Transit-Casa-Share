```python
data09 = pd.read_csv('E:/Transit-Casa-Alex/Output/Modeling/2009/Diff_Mapping_Data.csv',index_col = 0)
```

```python
data16 = pd.read_csv('E:/Transit-Casa-Alex/Output/Modeling/2016/Diff_Mapping_Data.csv',index_col = 0)
```

```python
merged = pd.merge(data09,data16,on = 'STOP_ID',how = 'outer',suffixes = ('_09','_16'))
```

```python
merged['MODELED_DIFF'] = merged.DIFF_AVG_RIDERSHIP_16 - merged.DIFF_AVG_RIDERSHIP_09
```

```python
merged['MODELED_PDIFF'] = ((merged.PDIFF_AVG_RIDERSHIP_16 - merged.PDIFF_AVG_RIDERSHIP_09)/merged.PDIFF_AVG_RIDERSHIP_09)*100
```

```python
merged.head()
```

```python
merged['MOD_OBS_DIFF_DIFF'] = merged.DIFF_AVG_RIDERSHIP_16 - merged.DIFF_AVG_RIDERSHIP_09
```

```python
merged['MOD_OBS_DIFF_PDIFF'] = ((merged.DIFF_AVG_RIDERSHIP_16 - merged.DIFF_AVG_RIDERSHIP_09)/merged.DIFF_AVG_RIDERSHIP_09)*100
```

```python
merged.head()
```

```python
merged['OBSERVED_DIFF'] = merged.AVG_RIDE_16 - merged.AVG_RIDE_09
```

```python
merged['GROWTH_DIFF'] = merged.MODELED_DIFF - merged.OBSERVED_DIFF
```

```python
merged['GROWTH_PDIFF'] = ((merged.MODELED_DIFF - merged.OBSERVED_DIFF)/merged.OBSERVED_DIFF)*100
```

```python
merged['GROWTH_DIFF'].mean
```

```python
merged.to_csv('Mapping_Data.csv')
```
