```python
scale_geo2 = scale_geo[['STOP_ID','BLKIDFP','BLKIDFP00','geometry','SPLIT_RATI']]
```

```python
def set_blockid(row):

    if row['BLKIDFP'] == np.nan: 
        row['BLKIDFP'] = row['BLKIDFP00']
    else:
        row['BLKIDFP'] = row['BLKIDFP']
    return row['BLKIDFP']  
```

```python
scale_geo2['BLKIDFP'] = scale_geo2.apply(lambda row: set_blockid(row),axis = 1)
```

```python
scale_geo2.BLKIDFP.count()
```

```python
scale_geo2.head()
```

```python
scale_geo2 = scale_geo2[['STOP_ID','BLKIDFP','SPLIT_RATI']]
```

```python
rename = ['STOP_ID','BLOCK_ID','SPLIT_RATIO']
```

```python
scale_geo2.columns = rename
```

```python
scale_geo2.head()
```

```python
split_grouped = scale_geo2[['SPLIT_RATIO','BLOCK_ID']].groupby(by = 'BLOCK_ID',as_index = False).sum()
```

```python
others_grouped = scale_geo2[['BLOCK_ID','STOP_ID']].groupby(by = 'BLOCK_ID',as_index = False).first()
```

```python
print(others_grouped.head())
print(len(others_grouped))
len(split_grouped)
```

```python
scale = pd.merge(others_grouped,split_grouped, on = 'BLOCK_ID',how = 'outer')
len(scale)
```
